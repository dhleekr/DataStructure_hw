### PROBLEM I - Linked List (15 POINTS)

Let us suppose that each node is linked to one of N nodes, where N is the number of nodes. The node can be linked to any other nodes including itself. Let's count the number of loops for ramdomly-linked nodes.

> A loop is defined as a set of nodes. A node can point out another one, making a path between them, and we can traverse the nodes along the path. If starting and ending nodes are the same, we call the set of nodes traversed as a loop. See the example shown below. The number of loop in this case is 3. For example, nodes 3 and 4 make a loop. Note that we can define a self-loop, e.g., the node 9. The node 5 is not a member of the loop composed of nodes 7, 8, and 6.
>
> $\texttt{point}$ is a list that records the name of the next node for each one. The node 0 is linked to the node 1, the node 1 is linked to the node 2, and so on in the example shown below. Then, values in the list of $\texttt{point}$ for the nodes 0 and 1 are 1 and 2, respectively, i.e., $\texttt{point[0]=1}$ and $\texttt{point[0]=2}$.
>
> $\texttt{nodes}$ is a list of N nodes, instances of the $\texttt{Node_}$ class. For the simplicity, we assume that a data field for each node is the index of the list itself. For example, the node 0 should be located in $\texttt{nodes[0]}$ and the corresponding data field is 0. A simple exmaple shown below shows the case of two nodes. Note that your algorithm should work for an arbitrary number of nodes. 
> ```python
self.nodes = [Node(0), Node(1)]
```
>
> **Example**
>
><img src = "https://drive.google.com/uc?id=1zXbNtmoVbIPQ3Vw-2LRbYNLwfWWxykVP" alt="add_attribute" width="400">
>
>> **Input**
> ```python
point = [1, 2, 3, 4, 3, 8, 7, 8, 6, 9]
```
>> **Output**
```
The number of loops:  3
4 -> 3
7 -> 8 -> 6
9
```

1. ```python
def link_nodes(self)
```

2. ```python
def cal_loops(self)
```

3. ```python
def print_loops(self)
```
> Print nodes and paths for each loop. Please refer to the sample codes for details.


### Implementation

````python
import random
from collections import OrderedDict

class Node_:
    def __init__(self, data_init):
        self.__data = data_init
        self.__next = None
        
    def get_data(self):
        return self.__data
    
    def get_next(self):
        return self.__next
    
    def set_data(self, new_alpha):
        self.__data = new_alpha
        
    def set_next(self, new_next):
        assert (isinstance(new_next, Node_) or new_next==None)
        self.__next = new_next
````

````python
class Problem1():
    def __init__(self, n, point):
        '''
        Parameters:
        n(int): The number of total nodes
        nodes(list): List of N nodes(Node_)
        point(list): Destinations of each node
        
        **DO NOT ADD ANY OTHER INPUT PARAMETERS**
        '''       
        self.nodes = [Node_(i) for i in range(n)]
        self.point = point
        self.n = n
        self.link_nodes()
        
    def link_nodes(self):
        '''
        Link each node to others using $\texttt{point}$(list, class variable)

        Returns:
        (nothing)
        '''
        for i in range(self.n):
            self.nodes[i].set_next(self.nodes[self.point[i]])
    def cal_loops(self):
        '''
        Compute the number of loops
        
        Returns:
        (int): return the number of loops
        '''
        count = 0
        nodes = list(self.nodes)
        passed = []
        while len(nodes) > 0:
            current = nodes[0]
            temp_list = []
            while current not in temp_list:
                if current not in passed:
                    is_loop = True
                    temp = current
                    temp_list.append(temp)
                    current = current.get_next()
                    nodes.remove(temp)
                else:
                    is_loop = False
                    break
            if is_loop:
                count += 1
            for i in temp_list:
                passed.append(i)
        return count
    
    def print_loops(self):
        '''
        Print nodes and paths for each loop.
        
        Returns:
        (nothing): Print out all loops
        '''
        nodes = list(self.nodes)
        passed = []
        while len(nodes) > 0:
            current = nodes[0]
            temp_list = []
            while current not in temp_list:
                if current not in passed:
                    is_loop = True
                    temp = current
                    current = current.get_next()
                    nodes.remove(temp)
                    temp_list.append(temp)
                else:
                    is_loop = False
                    break
            if is_loop:
                a = temp_list.index(current)
                for i in temp_list[a:]:
                    if temp_list.index(i) == len(temp_list)-1:
                        print(i.get_data())
                    else:
                        print(i.get_data(), '->', end = ' ')
            for i in temp_list:
                passed.append(i)
````

### Implementation

````python
'''
*** Do Not Touch(Modify) Here (class 'Node') ***
'''
import numpy as np
import pandas as pd

class Node(Node_):
    def __init__(self, data_init):
        Node_.__init__(self, data_init)
        self.__prev = None

    def get_prev(self):
        return self.__prev
    
    def set_prev(self, new_prev):
        assert (isinstance(new_prev, Node) or new_prev==None)
        self.__prev = new_prev
````

````python
class DoublyLinkedList:
    def __init__(self, attribute=None):
        self.head = None
        self.tail = None
        self.__attribute = attribute
    
    def __len__(self):
        '''
        Returns:
        (int): The number of nodes(entries)
       '''
        length = 0
        node = self.head
        while node != None:
            length += 1
            node = node.get_next()
        return length
        
    def is_empty(self):
        '''
        Returns:
        (bool): Check rather list has node(entry) or not
        '''
        
        return len(self) == 0
    
    def printpd(self, inverse=False):
        if self.data_list(inverse):
            data = self.data_list(inverse)
        else:
            return False
        
        dic = {}
        for i, attribute in enumerate(self.__attribute):
            dic[attribute] = np.array(data[i])
        df = pd.DataFrame(dic)
        display(df)
    
    def data_list(self, inverse=False):
        '''
        Parameters:
        inverse(bool): When False, print the table head to tail ordering
                       When True, print reversely
        
        Returns:
        (list or bool): Two-dimensional list to print out a table using pandas
                        if Doubly linked list is empty, return False
        '''
        
       
        if self.is_empty():
            return False
        
        data = np.zeros((len(self.__attribute),len(self)),dtype = 'object')
        
        if not inverse:
            temp = self.head
            for i in range(len(self.__attribute)):
                for j in range(len(self)):
                    data[i][j] = self.head.get_data()[i]
                    self.head = self.head.get_next()
                self.head = temp
                
        else:
            temp = self.tail
            for i in range(len(self.__attribute)):
                for j in range(len(self)):
                    data[i][j] = self.tail.get_data()[i]
                    self.tail = self.tail.get_prev()
                self.tail = temp
                
                
        return list(data)
    
    def search(self, attr, item):
        '''
        Parameters:
        attr(str): attribution of each node, where to check the item has or not 
        item(str or int): the target item to search

        Returns:
        (Node or bool): Return Node that we want to find.
                        Else return False to notice that method doesn't work well
        '''
        
        current = self.head
        found = False
        i = self.__attribute.index(attr)
        while not found and current != None:
            if current.get_data()[i] == item:
                found = True
            else:
                current = current.get_next()
        if found:
            return current.get_data()
        else:
            return False
        
    def add_entry(self, item, index=0):
        '''
        Parameters:
        item (Node): The node that we want to add from doubly linked list
        index (int): Index where to add entry
        
        Returns:
        (bool): Return True if remove given item properly.
                Else return False to notice that method doesn't work well
        '''
        
        # attribute랑 길이 안맞으면 안됨
        if len(self.__attribute) != len(item.get_data()): 
            return False
        
        # index 음수이면 (entry길이+1) 만큼 더해줌(index >= 0으로 만들어줌)
        if index < 0: 
            index = index + len(self) + 1
            
        # index 범위 안맞으면 안됨
        if index > len(self) or index < 0:
            return False
        
        # 마지막 꺼는 tail로
        
        if index == len(self):
            self.tail = item
        
        # index = 0 일때
        if index == 0:
            if self.head == None:
                self.head = item
                return True
            else:
                next_node = self.head
                self.head = item
                self.head.set_next(next_node)
                next_node.set_prev(self.head)
                return True
            
        # index 양수일때
        elif index > 0:
            current_index = 0
            current = self.head
            while current_index < index:
                if current:
                    pre_node = current
                    current = current.get_next()
                else:
                    break
                    
                current_index += 1
            if current_index == index:
                item.set_prev(pre_node)
                item.set_next(current)
                pre_node.set_next(item)
                if current:
                    current.set_prev(item)
                return True
            else:
                return False
       
                
        
    def add_attribute(self, attr, index=-1, default_value=None):
        '''
        Parameters:
        attr(str): attribute that we want to add from doubly linked list
        index(int): Index where to add attribute
        default_value(str or int):
        
        Returns:
        (bool): Return True if adding attribute is successed.
                Else return False to notice that method doesn't work well
        '''
        
        if index < 0:
            index = index + len(self.__attribute) + 1
        
        if index > len(self.__attribute) or index < 0:
            return False
            
        if attr in self.__attribute:
            return False
        
        current = self.head
        self.__attribute.insert(index, attr)
        while current != None:
            current.get_data().insert(index, default_value)
            current = current.get_next()
        return True
        
    def rm_entry(self, item):
        '''
        Parameters:
        item (Node): The node that we want to remove from doubly linked list

        Returns:
        (bool): Return True if remove given item properly.
                Else return False to notice that method doesn't work well
        '''
        
        if item == False:
            return False
        else:
            remove = False
            current = self.head
            for i in range(len(self)):
                if current.get_data() == item:
                    remove = True
                    pre_node = current.get_prev()
                    next_node = current.get_next()
                    pre_node.set_next(next_node)
                    next_node.set_prev(pre_node)
                else:
                    current = current.get_next()
            return remove
        
    def rm_attribute(self, attr):
        '''
        Parameters:
        att (str): The attribute that we want to remove from doubly linked list

        Returns:
        (bool): Return True if remove given attribute properly.
                Else return False to notice that method doesn't work well
        '''
    
        if attr not in self.__attribute:
            return False
        else:
            current = self.head
            i = self.__attribute.index(attr)
            self.__attribute.remove(attr)
            while current != None:
                del current.get_data()[i]
                current = current.get_next()
            return True
````

current = self.head 로 해주고 item을 찾았는지 여부를 나타내는 found = False로 해줍니다. 그리고 i에 attr의 self.\__attribute에서의 index를 넣어줍니다. 그래서 while문에 current가 한 칸씩 다음꺼로 가다가 None이 아닐때까지 & found가 False인 동안 반복을 해주면서 current의 값과 item의 값이 같아지면 found를 True로 바꿔주고 그렇지 않으면 get_next()를 통해 계속 다음꺼로 넘어갑니다. 그래서 found가 True로 while문이 끝나면 그 노드를 return해주고 False이면 return False를 해줍니다.      

## 5. add_entry     
먼저, attribute의 길이와 item의 길이가 안맞으면 False를 리턴해주고, index가 음수인 경우에는 (entry길이+1)만큼을 더해줘서 인덱스의 범위를 0≤i≤l 로 맞춰줬습니다(정상적 버범위의 index값이 들어온 경우). 그리고 index가 마지막인 경우에는 self.tail = item을 해줬습니다.     
1) index = 0 일때     
self.head = None이면 아무것도 없는 상태이므로 self.head = item을 해줍니다.    
self.head != None이면 원래 self.head에 들어있는걸 next_node에 넣어주고 item을 self.head에 넣어줍니다. 그리고 self.head와 next_node사이의 next, prev link를 해줍니다.       
2) index > 0 일때      
current_index = 0 을 해주고 current = self.head를 해줍니다. 그래서 parameter로 받은 index보다 current_index보다 작을동안 pre_node에는 current를 넣어주고 current를 다음 노드로 넘어가줍니다(current가 None이 아니면). current가 None이면 list가 끝까지 간거라서 break를 해줍니다. 그래서 current_index랑 index가 같으면 pre_node, current, item의 next, prev link를 해줍니다.      
정상적으로 작동할때에는 return True를 해줬고 그렇지 않을때에는 return False를 해줬습니다.       

## 6. add_attribute     
먼저, index가 음수인 경우에는 (attribute의 길이 + 1) 을 더해줘서 인덱스의 범위를 0≤i≤l 로 맞춰줬습니다(정상적인 범위의 index의 경우). 그리고 index의 범위가 맞지않으면 return False 를 해줬습니다. 그리고 attr가 self.\__attribute에 없으면 return False를 해줬습니다. current = self.head를 해주고 self.\__attribute에 index와 attr를 insert해줬습니다. 그리고 current가 None이 아닌 동안에 반복하는 while문을 만들어서 current가 계속 다음 노드로 넘어가면서 해당 attr의 index위치에 default value를 insert하는 작업을 해줬습니다. 그리고 마지막으로 정상적으로 작동하면 return True를 해줬습니다.       

## 7. rm_entry     
먼저 item == False 이면 return False를 해줬습니다.    
그리고 item이 정상적으로 들어온 경우에는 제거가 잘 됐는지의 여부인 remove를 만들어서 일단 False로 해놓고 current = self.head로 해놨습니다. 그래서 doubly linked list의 길이만큼 반복하면서 current가 처음부터 다음으로 계속 넘어가면서 item이랑 값이 같은지 여부를 판단하고 item이랑 같을때 remove를 True로 해주고 (pre_node, current, next_node) 이 순서에서 (pre_node, next_node)로 바꿔줍니다. 그래서 마지막으로 remove를 return 해줍니다. 제대로 제거됐다면 remove가 True이므로 True가 return 되고 그렇지 않다면 remove가 False 이므로 False가 return됩니다.       

## 8 .rm_attribute      
먼저, attr가 self.\__attribute에 없을경우 return False를 해줍니다.     
그리고 안에 제대로 있는경우 current = self.head로 해놓고 i에는 attr의 self.\__attribute에서의 index를 넣어줍니다. 그리고 attr를 self.\__attribute에서 없애줍니다. 그래서 current가 None이 아닐동안에 반복하는 while문을 만들어서 current가 self.head에서 계속해서 다음으로 넘어가면서 attr의 index위치에 있는 값들을 노드에서 전부 제거해줍니다. 제대로 작동할경우 return True를 해줍니다.

### PROBLEM III -  Recursion (5 POINTS)

> We can write a code to solve a 'recurrence relation', which could be solved more effectively using recursion. Write your code to solve the following problem <span style="color:red">using recursion</span>.
>
>>Suppose that a sequence of $\{a_n\}$, where $a_2=-1$, $a_3=2$. 
>>
>>The recurrence relation of the sequence is defined as follows: 
>>
>>$2a_{n+1}$ = $a_{n}$ + $a_{n+2}$ $(n=\cdots, -2, -1, 0,1,2,\cdots)$.
>>
>> Compute $\sum_{n=-5}^{10}a_n$.
>
> Here, we do not provide an answer. Please solve the problem by yourself, and check whether your solution is the same or not as the one obtained by you code. 


### Implementation

````python
def a(x):
    if x == 2:
        return -1
    elif x == 3:
        return 2
    elif x > 3:
        return 2*a(x-1)-a(x-2)
    else:
        return 2*a(x+1)-a(x+2)

sum = 0 
for i in range(-5,11):
    sum = sum + a(i)

print(sum)
````

### Describe your answer: 

먼저, a2와 a3은 주어진 값이므로 주어진 값을 return합니다.
그리고 n > 3 이상을 구할때는 a2 a3 기준으로 n이 큰 값을 구하는 것이기 때문에 점화식 부분에서 a<sub>n</sub> = 2a<sub>n-1</sub> - a<sub>n-2</sub>으로 다음 값을 구할 수 있게 recursion을 이용했습니다. 그리고 n < 2 인 경우(else부분) 점화식 부분에서 a<sub>n</sub> = 2a<sub>n+1</sub> - a<sub>n+2</sub>을 으로 a<sub>1</sub>, a<sub>0</sub>,...을 recursion을 이용해서 구할 수 있게했습니다. 그래서 sum = 0으로 해놓고 함수를 이용해서 for문으로 -5부터 10까지 a<sub>n</sub>을 return해서 sum에 계속 더해줌으로써 sum을 구했습니다.

### PROBLEM IV - Recursion (Advanced) (30 POINTS)

> Write a code to complete the $\texttt{Maze}$ class. The maze map problem aims at finding possible paths starting from the blank(o) to the finish point (F). Please see the sample code provided for details. 
>
>The maze map can have three components: Blank (o), Wall (+), and Finish (F). We can start any blanks (o) to find the paths. 
>
>To this  end, we have four choices: left, right, up, and down. We are not allowed to pass through the Wall (+), and can move toward either Blank (o) or Finish (F) only. We are not allowed to go outside of the given maze map. 



#### **Instruction**

> 1. Complete search method
```python
def search(self, y, x)
```
>> Determine whether we can find any paths starting from (y, x) for the the maze map.
>>
>> You <span style="color:red">must use recursion.</span> 
>> You can use additional input parameters if you need.
>
> 2. Complete solve method
```python
def solve(self, y, x)
```
>> Find and print a single line path for the maze map. You can not revisit the position previously visited. In other words, the line of the path must not overlap itself when drawn in a single line. It does not need to be the shortest path. Add marker(S) at starting point. See the example provided for your understand.
>> You can use additional input parameters if you need.


> Make sure that we will test your code on various maze maps.
>
> Do not directly use given maps or distort them. That is, after the execution of search or solve methods, the map should be the same as before calling the methods. 

### Implementation

````python
import numpy as np

class Maze():
    def __init__(self, grid):
        '''
        Parameters:
        grid (list): The maze map to solve
        
        **YOU CAN NOT ADD ANY OTHER INPUT PARAMETERS **
       '''
        self.data = np.array(grid)
        
    def search(self, y, x):
        '''
        Parameters:
        y (int): Start location of y
        x (int): Start location of x
        
        **The starting point is (y,x)
        **The coordinates for the left-top, left-bottom, right-top, and right-bottom are 
        ** (0,0), (N,0), (0,N), (N,N), respectively, for the maze map of size N by N.
        **YOU CAN ADD ANY OTHER INPUT PARAMETERS IF YOU WANT**
        **DESCRIBE THE ADDITIONAL PARAMETERS IN YOUR DESCRIPTION**

        Returns:
        (bool): If maze has solution or not, return True or False respectively

       '''
        if self.data[y][x] == 'F':
            return True
        if self.data[y][x] == '+':
            return False
        if self.data[y][x] == 'x':
            return False
        
        self.data[y][x] = 'x'
        
        if y < self.data.shape[0]-1 and self.search(y+1,x):
            self.data[y][x] = "↓"
            return True
        elif x>0 and self.search(y,x-1):
            self.data[y][x] = "←"
            return True
        elif y>0 and self.search(y-1,x):
            self.data[y][x] = '↑'
            return True
        elif x < self.data.shape[1]-1 and self.search(y,x+1):
            self.data[y][x] = '→'
            return True
        return False
        
    def solve(self, y, x):
        '''
        Parameters:
        y (int): Start location of y
        x (int): Start location of x
        
        **The starting point is (y,x)
        **The coordinates for the left-top, left-bottom, right-top, and right-bottom are 
        ** (0,0), (N,0), (0,N), (N,N), respectively, for the maze map of size N by N.
        **YOU CAN ADD ANY OTHER INPUT PARAMETERS IF YOU WANT**
        **DESCRIBE THE ADDITIONAL PARAMETERS IN YOUR DESCRIPTION**

        Returns:
        (nothing): Print out a single path solution

       '''
        if self.search(y, x):
            print("Solution")
            self.data[y][x] = 'S'
            for i in range(self.data.shape[1]):
                for j in range(self.data.shape[0]):
                    if self.data[j][i] == 'x':
                        self.data[j][i] = 'o'
            print(self.data)
        else:
            print('There are NO solutions')
````

````python
for i, maze_map in enumerate(maze_map_list):  
    maze = Maze(maze_map)
    print(i+1, '-1')
    maze.solve(0,0)
    print(i+1, '-2')
    maze.solve(4,0)
    print()
````
