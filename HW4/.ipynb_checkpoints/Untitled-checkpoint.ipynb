```python
class HashTable:
    def __init__(self, size, max_chain):
        self.size = size
        self.max_chain = max_chain
        self.slots = [[] for _ in range(size)]
        self.data = [[] for _ in range(size)]
        
    def __len__(self): #item의 개수
        length = 0
        for i in range(self.size):
            length += len(self.slots[i])
        return length
    
    def hash_func(self, key):
        return key % self.size
    
    def rehash_func(self, key, cnt):
        return (key + cnt**2) % self.size
    
    def put(self, key, data):
        hash_value = self.hash_func(key)
        i = 0
        complete = False
    
        while not complete:
            if key in self.slots[hash_value]:
                    self.data[hash_value][self.slots[hash_value].index(key)] = data
                    complete = True
            else:
                if len(self.slots[hash_value]) < self.max_chain:
                    self.slots[hash_value].append(key)
                    self.data[hash_value].append(data)
                    complete = True
                else:
                    i += 1
                    hash_value = self.rehash_func(key, i)
        return i
          
    def get(self, key):
        start_slot = self.hash_func(key)
        data = None
        stop = False
        found = False
        i = 1
        hash_value = start_slot
        while self.slots[hash_value] != None and not found and not stop:
            if self.slots[hash_value].count(key) != 0:
                index = self.slots[hash_value].index(key)
                data = self.data[hash_value][index]
                found = True
            else:
                hash_value = self.rehash_func(key, i)
                if hash_value == start_slot:
                    stop = True  
            i += 1
        if stop:
            print('Key does not exist')
        else:
            return data
    
    def remove(self, data):
        found = False
        for i in range(self.size):
            if self.data[i].count(data) > 0:
                found = True
                for j in range(len(self.data[i])-1,-1,-1):
                    if self.data[i][j] == data:
                        del self.slots[i][j]
                        del self.data[i][j]
        
        if not found:
            print('Data does not exist')
                
    def __getitem__(self, key):
        return self.get(key)
    
    def __setitem__(self, key, data):
        self.put(key, data)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import time

# DO NOT modify
np.random.seed(0)
random_keys = np.random.choice(10000, 1000, replace = False)
random_data = np.random.choice(10000, 1000)

# DO NOT modify this class
class random_input:
    def __init__(self):
        self.cnt = -1
        
    def __call__(self):
        self.cnt += 1
        return random_keys[self.cnt], random_data[self.cnt]

#############
def load_factor(table):
    count = 0
    for i in range(len(table.slots)):
        if table.slots[i] != []:
            count += 1
    factor = len(table)/ count
    return factor

table = HashTable(size = 100, max_chain = 10)

a = random_input()
lf = 0
factor = []
rehash = []

while lf < 9.5:
    temp = a()
    rehash.append(table.put(temp[0], temp[1]))
    lf = load_factor(table)
    factor.append(lf)

plt.plot(factor,rehash)
plt.xlabel('load factor')
plt.ylabel('# of calls on rehashing function')
plt.yticks(range(0,max(rehash),4))
plt.show()
#############
```
