### # PROBLEM I (20 Points)

(a) Build a **sequential search** and **binary search** algorithm to find integers while satisfying following conditions:  **[15 points]**

> 1. Return list of **all** the indices that contain integers which is exactly same as $num$.  
>  
>
> 2. If none of integers in the list satisfy above conditions, simply return an empty list.

#### Implementation Rules :  

> 1. **DO NOT use** `list.index()` method, or numpy module.
2. Assume that `search_list` is sorted in descending order, but can have overlapping elements.
3. Use recursion when implementing binary search.

```python
def sequential_search(search_list, num):
    index = 0
    res = []
    while index < len(search_list):
        if search_list[index] == num:
            res.append(index)
        index += 1
    return res


def binary_search(search_list, num):
    result = []
    if len(search_list) == 0:
        return []
    else:
        mid = len(search_list) // 2
    
    if search_list[mid] == num:
        left = mid
        right = mid
        result.append(mid)

        while(left > 0):
            if search_list[left - 1] == num:
                left -= 1
                result.insert(0, left)
            else:
                break

        while(right < len(search_list) - 1):
            if search_list[right + 1] == num:
                right += 1
                result.append(right)
            else:
                break
      
        return result
      
    elif search_list[mid] < num:
        return binary_search(search_list[: mid], num)
    else:
        temp = binary_search(search_list[mid+1 :], num)
        i = 0
        if len(temp):
            while(len(temp) > i):
                temp[i] += mid + 1
                i += 1
        return temp
```

```python
# DO NOT modify this code
search_list = [5, 4, 4, 3, 3, 2, 1]
print(sequential_search(search_list, 3))
print(binary_search(search_list, 4))
```

### # PROBLEM II (20 Points)

The methods of LinkedList is specified as follows:
* <code>\_\_init\_\_()</code>: constructs an empty LinkedLIst.
* <code>\_\_len\_\_()</code>: returns the number of items in the list.
* <code>insert(item)</code>: inserts $item$ in the list.
* <code>search(x)</code>: returns all the indices of Nodes satisfying $Node.__data = x$. The index of a Node starts from 0, which is the head of the list.
* <code>search_diff(interval)</code>: returns all the indices of Nodes where the $data$ part of next Node is higher than that of current Node by a gap of $interval$.
* <code>\_\_le\_\_(self, other)</code>: returns boolean value whether the LinkedList is $smaller\ than$ or $equal\ to$ other LinkedList.  
A LinkedList $list1$ is deemed to be $smaller\ than$ or $equal\ to$ LinkedList $list2$ if they satisfy following conditions:
> 1. The length of $list1$ is smaller than or equal to that of $list2$.
> 2. $list2$ contains all the elements of $list1$.

Complete LinkedList specified above.

**You can define other methods for implementation. But DO NOT modify any predefined methods inside the class.  
Just code at sections marked as `# Code here #`.**

```python
# DO NOT change this code
class Node:
    def __init__(self, data):
        self.__data = data
        self.__next = None
        
    def setdata(self, _data):
        self.__data = _data
        
    def setnext(self, _next):
        self.__next = _next
        
    def getdata(self):
        return self.__data
    
    def getnext(self):
        return self.__next
```

```python
class LinkedList:
    def __init__(self):
        self.__head = None
        self.length = 0
        
    def __len__(self):
        return self.length
        
    def insert(self, item):
        temp = Node(item)
        temp.setnext(self.__head)
        self.__head = temp
        self.length += 1
        
    def search(self, x):
        res = []
        current = self.__head
        index = 0
        while current != None:
            if current.getdata() == x:
                res.append(index)
            current = current.getnext()
            index += 1
        return res
    
    def search_diff(self, interval):
        res = []
        current = self.__head
        if current != None:
            next_node = current.getnext()
        index = 0
        while current != None and next_node != None:
            if next_node.getdata() - current.getdata() == interval:
                res.append(index)
            current = current.getnext()
            next_node = next_node.getnext()
            index += 1
        return res
    
    def __le__(self, other):
        if len(self) > len(other):
            return False
        else:
            current = self.__head
            while current != None:
                if len(other.search(current.getdata())) == 0: #있는지 없는지
                    return False
                else:
                    current = current.getnext()
            return True
                    
        
```

```python
# DO NOT modify this code
list1 = LinkedList()
list2 = LinkedList()

list1.insert(3)
list1.insert(2)
list1.insert(5)
list1.insert(3) #list1 = [3,5,2,3]

list2.insert(1)
list2.insert(2)
list2.insert(5)
list2.insert(3)
list2.insert(4) #list2 = [4,3,5,2,1]

print(list1.search(3))
print(list2.search_diff(-1))
print(list1 <= list2)
```

### Description:

### 1. search
res 라는 결과로 출력할 리스트를 만들고 current = self.\_\_head 로 해놓고 index = 0 으로 해놓고 current != None 일때까지(linkedlist 끝) 반복하는 while문을 만들었습니다. 그래서 current.getdata()와 x 가 값이 같은지를 확인하고 같으면 res.append(index)를 했습니다. 그리고나서 index += 1을 해주고, current는 다음노드로 넘어가게 했습니다. 그래서 끝나면 return res 를 통해서 index의 리스트를 출력했습니다.       

### 2. search_diff      
search와 거의 동일한 방식인데 next_node = current.getnext()로 current의 다음 노드로 해주고 next_node.getdata() - current.getdata() == interval 일때 res.append(index)를 해줘서 조건을 만족할때만 index가 추가되게 해줬습니다.       

### 3. \_\_le\_\_     
먼저, len(self)가 len(other)보다 크면 안되기때문에 처음 if문에서 self가 더 큰 경우 return False로 해줬습니다.      
self와 other의 길이 조건을 만족한 경우에는 이제 current = self.\_\_head로 해주고 while current != None: (current노드가 다음노드로 계속 넘어가다가 끝날때까지 반복)을 해줬습니다.     
그 안에서 other안에 current의 값이 있는지 찾기 위해서 search 함수를 이용했습니다. 그래서 len(other.search(current.getdata())) == 0 이면 other안에 current값이 없다는 뜻이기 때문에 return False를 해줬습니다.      
other안에 current값이 있는경우, current를 다음 노드로 넘어가게 해줬습니다.       
그래서 while문이 끝날때까지(self의 모든 값을 검토) return False가 안됐다면 other에 self의 모든 원소들이 포함되므로 return True를 해줬습니다.

### # PROBLEM III (15 Points)

Implement a HashTable named **Naive_HashTable** satisfying following conditions:

> 1. The HashTable is initialized with $size$, where $size$ refers to the size of HashTable.
>
>
> 2. The HashTable prevents collision only by **chaining**. Chaining allows many items to exist at the same location in the HashTable. 

Methods of **Naive_HashTable** is specified as follows :  
* <code>\_\_init\_\_(size, max_chain)</code>: constructs an empty HashTable.
* <code>hash_func(key)</code>: returns hash value of given $key$.
* <code>put(key, data)</code>: When a item composed of $(key, data)$ is put to the HashTable, the HashTable first computes the location to store the item by `hash_func(key)` method. Then, the item is added to the list, at the hashed location.
* <code>get(key)</code>: returns $data$ stored in HashTable under given $key$. If $key$ does not exist, print relevant error message.

**You can define other methods for implementation. But DO NOT modify any predefined methods inside the class.  
Just code at sections marked as `# Code here #`.**

```python
class Naive_HashTable:
    def __init__(self, size):
        self.size = size
        self.slots = [[] for _ in range(size)]
        self.data = [[] for _ in range(size)]

    def hash_func(self, key):
        return key % self.size
    
    def put(self, key, data):
        hash_value = self.hash_func(key)
        if key in self.slots[hash_value]:
            self.data[hash_value][self.slots[hash_value].index(key)] = data
        else:
            self.slots[hash_value].append(key)
            self.data[hash_value].append(data)

    def get(self, key):
        hash_value = self.hash_func(key)
        if len(self.slots[hash_value]) == 0:
            print('No corresponding key')
        else:
            if self.slots[hash_value].count(key) == 0:
                print('No corresponding key') 
            else:
                return self.data[hash_value][self.slots[hash_value].index(key)]
```

```python
# DO NOT modify this code
naive = Naive_HashTable(5)

naive.put(5, 'a')
naive.put(2, 'b')
naive.put(7, 'c')

print(naive.get(2))
print(naive.get(7))
print(naive.get(3))
```

### Description:

### 1. put
먼저, hash_func을 이용해서 hash_value를 만들어줬습니다.    
key가 이미 self.slots[hash\_value]에 들어있으면 중복되면 안되므로 그것에 해당하는 data의 값을 replace 해줬습니다.       
key가 self.slots[hash\_value]에 없으면 self.slots[hash\_value]와 self.data[hash\_value]에 각각 key와 data를 append 해줬습니다.        

### 2. get       
먼저, hash_func을 이용해서 hash_value를 만들어줬습니다.      
그리고 만약 hash_value 위치에 아무것도 없는경우에는 hashtable에 input으로 들어온 key가 없는 것이므로 print('No corresponding key')를 해줬습니다.     
그리고 hash_value 위치에 아무것도 없지 않은 경우에, key값이 있는지 없는지를 list.count() method를 사용해서 0인지 아닌지로 판단을 했고, self.slots[hash_value].count(key) == 0 이면 key값에 해당하는 hash_value는 있지만 key는 없는 것이므로 print('No corresponding key')를 해줬습니다.       
그리고 key값도 있는 상황이라면 self.slots[hash\_value].index(key)를 통해서 해당하는 hash_value의 리스트 안에 key index를 얻어서 self.data[hash_value] 리스트에서 해당하는 index의 data를 return 해줬습니다.

### # PROBLEM IV (45 Points)

(a) Implement a **HashTable** satisfying following conditions: **[35 points]**
> 1. The HashTable is initialized with $size$ and $max\_chain$.  
$size$ refers to the size of HashTable, and $max\_chain$ would be explained in condition 3.  
>
>
> 2. The HashTable uses **quadratic probing**.  
That is, if the first hash value is $h$, successive values would be $h + 1$, $h + 4$, ..., and so on.  
>
>
> 3. The HashTable also uses **chaining**. But one location can contain up to $max\_chain$ items.  
>
>
> 4. If an item is put to HashTable, it is stored to firstly-hashed value when the location has less than $max\_chain$ items in it.  
Otherwise, it finds other locations by quadratic probing.

Methods of **HashTable** are specified as follows:  
* <code>\_\_init\_\_(size, max_chain)</code>: constructs an empty HashTable.
* <code>\_\_len\_\_()</code>: returns total number of items in the HashTable.
* <code>hash_func(key)</code>: returns first hash value of given $key$.
* <code>rehash_func(key, cnt)</code>: returns $(cnt + 1)$-th hash value of given $key$, satisfying condition 2.
* <code>put(key, data)</code>: puts $data$ in HashTable according to its given $key$, and returns the number of calls on `rehash_func()` method  
when putting $(key, data)$ into the HashTable.  
* <code>get(key)</code>: returns $data$ stored in HashTable under given $key$. If $key$ does not exist, print relevant error message.
* <code>remove(data)</code>: deletes $data$ and corresponding $key$ in HashTable. If $data$ does not exist, print relevant error message.  
If the HashTable has more than one $data$, delete them all.

**When implementing `put()` and `get()` method, USE `hash_func()` / `rehash_func()` method to access keys rather than exhaustive search.**  
**You can define other methods for implementation. But DO NOT modify any predefined methods inside the class.  
Just code at sections marked as `# Code here #`.**

### Answer:

```python
class HashTable:
    def __init__(self, size, max_chain):
        self.size = size
        self.max_chain = max_chain
        self.slots = [[] for _ in range(size)]
        self.data = [[] for _ in range(size)]
        
    def __len__(self): #item의 개수
        length = 0
        for i in range(self.size):
            length += len(self.slots[i])
        return length
    
    def hash_func(self, key):
        return key % self.size
    
    def rehash_func(self, key, cnt):
        return (key + cnt**2) % self.size
    
    def put(self, key, data):
        hash_value = self.hash_func(key)
        i = 0
        complete = False
    
        while not complete:
            if key in self.slots[hash_value]:
                    self.data[hash_value][self.slots[hash_value].index(key)] = data
                    complete = True
            else:
                if len(self.slots[hash_value]) < self.max_chain:
                    self.slots[hash_value].append(key)
                    self.data[hash_value].append(data)
                    complete = True
                else:
                    i += 1
                    hash_value = self.rehash_func(key, i)
        return i
          
    def get(self, key):
        start_slot = self.hash_func(key)
        data = None
        stop = False
        found = False
        i = 1
        hash_value = start_slot
        while self.slots[hash_value] != None and not found and not stop:
            if self.slots[hash_value].count(key) != 0:
                index = self.slots[hash_value].index(key)
                data = self.data[hash_value][index]
                found = True
            else:
                hash_value = self.rehash_func(key, i)
                if hash_value == start_slot:
                    stop = True  
            i += 1
        if stop:
            print('There is no corresponding key.')
        else:
            return data
    
    def remove(self, data):
        found = False
        for i in range(self.size):
            if self.data[i].count(data) > 0:
                found = True
                for j in range(len(self.data[i])-1,-1,-1):
                    if self.data[i][j] == data:
                        del self.slots[i][j]
                        del self.data[i][j]
        
        if not found:
            print('There is no corresponding data.')
                
    def __getitem__(self, key):
        return self.get(key)
    
    def __setitem__(self, key, data):
        self.put(key, data)
```

```python
# This code is for validation; DO NOT change this code
table = HashTable(size = 7, max_chain = 2)
key = list(range(7))
data = ['a', 'b', 'c', 'd', 'e', 'f', 'g']

for i, j in zip(key, data):
    table[i] = j
    
print(table[4])
key = list(range(7, 14))
for i, j in zip(key, data):
    table[i] = j
table.remove('c')
table.remove('f')
table.put(1, 'h')
table.put(13, 'i')
print(table.data)
print('-' * 40)
table.remove('d')
print(table.put(15, 'c'))
print(table.slots)
print('-' * 40)
table.get(21) # Error message
table.remove('x') # Error message
print('-' * 40)
print(len(table))
table[5] = 'a'
table.remove('a')
print(table[15])
```

### Description:

### 1. \_\_len\_\_ 
length = 0 으로 해두고, self.slots의 모든 부분을 돌면서 각 index에 있는 리스트의 길이를 length에 더 해줬습니다.     

### 2. rehash_func     
Quadratic probing 이므로 key, cnt를 입력 받으면 cnt의 제곱을 더해주고 size로 나눈 나머지를 return 해줬습니다.       
 
### 3. put      
hash_value를 self.hash_func(key)를 통해서 먼저 구해줍니다. 그리고 i = 0 으로 해두고 complete = False로 한 후에 complete이 False인 동안 반복되는 while문을 만들었습니다.   
그래서 만약 key가 self.slots[hash_value]에 있는 경우 data를 replace해주고 complete = True로 해줬습니다.      
만약, key가 self.slots[hash_value]에 없는 경우 len(self.slots[hash_value])가 self.max_chain보다 작은 경우와 큰 경우로 나눴습니다. 그래서 작은 경우에는 key와 data를 그냥 append 해주고 complete = True로 해줬습니다. 큰 경우에는 i에 1을 더해주고 rehashing을 해줘서 hash_value를 새로 구하고 위의 과정을 반복해줬습니다. 그래서 결과적으로 i의 값이 rehash_func 몇 번 했는지 값이므로 return i 를 해줬습니다.     

### 4. get     
start_slot = self.hash_func(key), data = None, stop = False, found = False, i = 1, hash_value = start_slot으로 해줬습니다. 그리고 self.slots[hash_value]가 안비어있고 found = False이고 stop = False인 동안 반복되는 while문을 만들어줬습니다. 만약 self.slots[hash_value].count(key) != 0 이면 self.slots[hash_value] 안에 key가 들어있는 것이므로 index = self.slots[hash_value].index(key) 로 key의 index를 찾아서 data = self.data[hash_value][index]로 해서 data 값을 찾고 found = True로 바꿔줍니다. 만약 key가 안들어있으면 hash_value를 rehash해주고 만약 rehash해준 hash_value가 start_slot이랑 같은 경우에는 모든 경우를 다 돌아본 경우이므로 stop = True를 해줬습니다. 그리고 i += 1을 해줍니다. 그래서 stop이 True면 key가 없는거니까 print('There is no corresponding key.')을 출력해주고 아니면 return data를 해줬습니다.       
 
### 5. remove      
먼저 found = False로 해주고 첫번째 for문으로 self.data[i].count(data) 로 data가 들어있는지 없는지를 판단을 해주고 들어있으면 found = True로 해주고 두번째 for문으로 list의 마지막 값부터해서 data랑 같은지 판단해서 같은면 그 위치의 self.slots와 self.data의 값을 지워줬습니다.      
그래서 for문이 끝난경우 found가 False인 경우 data가 없었던 경우이므로 print('There is no corresponding data.')를 해줬습니다.
