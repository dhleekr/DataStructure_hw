```python
# DO NOT change this code
import numpy as np

def random_numbers(num = 1000):
    np.random.seed(1)
    num_even = num // 2
    numbers = np.random.choice(5000, num, replace = False) * 2
    idx = np.random.choice(num, num_even, replace = False)
    numbers[idx] += 1
    return numbers
```

```python
# Code for (a)

def bubble_sort(x): # 2 pts
    for i in range(len(x)-1, 0, -1):
        for j in range(i):
            if x[j] > x[j+1]:
                temp = x[j]
                x[j] = x[j+1]
                x[j+1] = temp
    
def shell_sort(x): # 3 pts
    gap = len(x) // 2
     # loop over the gaps
    while gap > 0:
         # do the insertion sort
        for i in range(gap, len(x)):
            val = x[i]
            j = i
            while j >= gap and x[j - gap] < val:
                x[j] = x[j - gap]
                j -= gap
            x[j] = val
        gap //= 2
```
